# Chabelo

```python
import random as rand
```

```python
# numero de experimentos
experiments = 1000

# La forma de resolver este problema fue a travs del uso de listas
# cada casilla de una lista representa una iteracion de la simulacion
# Existen 3 posibles puertas a escoger [0,1,2]


# Numero de puerta en la que esta el premio
prize = [rand.randint(0,2) for i in range(experiments)]

# Seleccion inicial de puerta
my_choice=[rand.randint(0,2) for i in range(experiments)]

# Puerta que abre chabelo
chabelo_open=[]

# Seleccion de puerta despues de que chabelo abre una puerta, siempre cambiamos de opcion.
new_choice=[]

# Contador para saber cuantas veces ganamos el juego
cont = 0.0


for i in range (0, experiments):
    
    chabelo_options=[]
    
    # agregamos a una lista las posibles puertas que puede abrir chabelo en la iteracion i
    for j in range(0,3):
        if(prize[i] != j and my_choice[i] != j):
            chabelo_options.append(j)

    # Chabelo abre una puerta elegida aleatoriamente entre sus opciones
    random = rand.randint(0,len(chabelo_options)-1)
    chabelo_open.append(chabelo_options[random])
    
    # Hacemos el cambio de puerta elegida
    if (random == 0):
        complemento = 1
    else:
        complemento = 0
    if(prize[i] == my_choice[i]):
        new_choice.append(chabelo_options[complemento])
    else:
        new_choice.append(prize[i])
    
    # Verificamos si gano el juego despues del cambio
    if (new_choice[i] == prize[i]):
        cont = cont + 1

        
        
        
        
        
        
# Visualisacion

print "Porcentaje de Ganar"
print cont/experiments

'''

for i in range (0, experiments):
    print ""
    print ""
    print "Iteracion:"
    print ""
    print "prize: "
    print prize[i]
    print "choice: "
    print my_choice[i]
    print "chabelo: "
    print chabelo_open[i]
    print "new choice: "
    print new_choice[i]  

'''



        
```
